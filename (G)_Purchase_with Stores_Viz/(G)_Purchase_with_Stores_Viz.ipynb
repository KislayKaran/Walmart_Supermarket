# Showing the Consumer Location (Blue Dots) and their nearest walmart stores(Red Dots) with the help of Map Density.


```python
 import pandas as pd
```


We will now be finding the closest store from the user's address based on
user address longitude/latitude and store longitude/latitude in close proximity.

```python
purchase_history_5= pd.read_csv('https://raw.githubusercontent.com/KislayKaran/Walmart_Supermarket/main/Required_CSV_files/13_c_purchase_history_5_states_%5B(created_csv)%20using_faker%5D.csv')
```

```python
all_stores= pd.read_csv('https://raw.githubusercontent.com/KislayKaran/Walmart_Supermarket/main/Required_CSV_files/12_c_master_store_data_(created_csv).csv')
```

```python
purchase_history_5=purchase_history_5[['family_id','super_market_uuid']]
```

```python
purchase_history_5.head()
```

We need to generate account , need to group by and generate the account and the total no of items they have bought.


```python
final_purchase= purchase_history_5.groupby(['family_id','super_market_uuid']).size().to_frame(name='total').reset_index()
final_purchase
```

```python
df_address= pd.read_csv('https://raw.githubusercontent.com/KislayKaran/Walmart_Supermarket/main/Required_CSV_files/06_c_user_address_%5B(created_csv)%20using_faker%5D.csv')
```

Merge with LEFT Join

```python
add_temp=pd.merge(final_purchase,df_address, how = 'left', on= 'family_id')
add_temp
```

```python
purchase_df=add_temp[['family_id','super_market_uuid','total','city','state','longitude','latitude']]
purchase_df
```
