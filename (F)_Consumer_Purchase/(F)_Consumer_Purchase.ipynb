We will now be finding the closest store from the user's address based on
user address longitude/latitude and store longitude/latitude in close proximity.
 Basically we need to know that from which particular Walmart store did the User consumed the products.

```python
stores= pd.read_csv('https://raw.githubusercontent.com/KislayKaran/Walmart_Supermarket/main/Required_CSV_files/12_c_master_store_data_(created_csv).csv')
stores
```

```python
PRECISION=1
```

```python
stores['latitude']= stores.latitude.round(PRECISION)
stores['longitude']= stores.longitude.round(PRECISION)
```

```python
def get_closest_super_market(family_longitude,family_latitude):
  supermarket_id= None
  local_longitude=round(family_longitude,1)
  local_latitude=round(family_latitude,1)
  stores_select_long = stores[stores['longitude']==local_longitude]
  stores_long_lati = stores_select_long[stores_select_long['latitude']== local_latitude]
  if len(stores_long_lati)>0:
    selected_store= stores_long_lati.sample(1).to_dict(orient='records')[0]
    supermarket_id= selected_store['store_id']

  return supermarket_id

```

we reduced the family latitude and longitude to single precision and also we reduced the store latittude and longitude and then we match it equal.

Then out of the equal precisions we then randomly select one particular store and then try attaching it to that user address. Because there might two different stores with the same precised co-ordinates in the same area.

```python
 # Creating Purchase History for the entire United States.......
purchase_history= []
years=[2020,2021]
for year in years:
  for month in range(12):
    firstday_month= datetime(year, month+1, 1)
    lastday_month= firstday_month + pd.offsets.MonthEnd(1)
    for index, family in user_add.iterrows():
      family_id= family['family_id']
      family_longitude= family['longitude']
      family_latitude= family['latitude']
      # Generating Random shopping visits in the selected month
      for i in range(random.randrange(1,5)):
        # Getting closet super _market uuid
        super_market_uuid = get_closest_super_market(family_longitude, family_latitude)
        if super_market_uuid is not None:
          # Generating Order Number
          order_number = faker.pystr_format(string_format="???-######")
          purchase_date = faker.date_time_between_dates(firstday_month, lastday_month)
          # Random no. of items purchased in the given order
          items_in_order = random.randrange(1,15)
          for prod_index, selected_prod in products.sample(items_in_order).iterrows():
            order_info = {'order_number': order_number, 'purchase_date': purchase_date, 'family_id': family_id, 'order_total': items_in_order,
                          'qty': random.randrange(1,5), 'super_market_uuid':super_market_uuid , 'product_id': selected_prod['product_id'], 'product_name': selected_prod['product_name'],
                          'price': round(random.uniform(1.0, 120.0), 2)}
            purchase_history.append(order_info)               
```
