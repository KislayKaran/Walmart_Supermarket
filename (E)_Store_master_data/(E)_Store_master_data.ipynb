```python
from datetime import datetime
import uuid
import random
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/KislayKaran/Walmart_Supermarket/main/Required_CSV_files/01_d_walmart_stores_2018_11_06_(downloaded_csv).csv')
df.head()
```

Adding the Custom ID to the each Walmart Store (UUID) along with the selected Fields

```python
df=df[['name','street_address','city','state','latitude','longitude']]
df.head()
```

```python
df['store_id'] = df.apply(lambda _: uuid.uuid4(), axis=1)
```

```python
df.to_csv('12_c_master_store_data_(created_csv).csv')
```
